```python
#recursive function
def recursive(n):
  if n<=1:
    return n
  return n + recursive(n-1)
n=5
print(recursive(n))
```

```python
def rec_fib(n):
  if n<=1:
    return(n)
  else:
    return(rec_fib(n-1) + rec_fib(n-2))
n=5
for i in range(n):
  print(rec_fib(i))
```

```python
def fact(n):
  if n<=1:
    return(n)
  else:
    return(n*fact(n-1))
n=5
print(fact(n))
```

```python
#generator fun
def gen_num(n):
  v=0
  while v<n:
    yield v
    v+=1
for v in gen_num(100):
    print(v)
```

```python
def calc_string(x):
  lower=0
  upper=0
  for i in x:
    if i.isupper():
      upper+=1
    else:
      lower+=1
      return lower
      return upper
print(calc_string("SeConD dAy of tHe wEEk"))
```
